## 4. Check for Duplication, then drop them

```python
## check duplicates in Country column 
country_Duplication = Data_Frame2015["Country"].duplicated()
print("Number of duplicates in the 'Country' column:", country_Duplication.sum())
```

```python
# Define function to get unique regions
def get_unique_regions():
    unique_regions = Data_Frame2015["Region"].unique().tolist()
    return unique_regions
unique_regions = get_unique_regions()
print("Number of unique regions:", len(unique_regions))

# Call function to get unique regions and print them using a loop
unique_regions = get_unique_regions()
print("Unique regions:")
for region in unique_regions:
    print(region)
```

```python
# Check for duplicates in Region column
region_Duplication = Data_Frame2015["Region"].duplicated()
print("Number of duplicates in the 'Region' column:", region_Duplication.sum())
```

```python
# Count duplicates for each region
dupli = Data_Frame2015[Data_Frame2015.duplicated(subset=["Region"], keep=False)]
dupli_count = dupli.groupby("Region").size()

# Print duplicates count for each region
print("Duplicates count for each region:")
print(dupli_count)
```

```python
# Drop duplicates in Region column
Data_Frame2015.drop_duplicates(subset=["Region"], inplace=True)
```

```python
# check that all the duplicates are dropped 

region_Duplication_dr = Data_Frame2015["Region"].duplicated()
print("Number of duplicates in the 'Region' column:", region_Duplication_dr.sum())
```

## 2. Creating the heatmap to visualize the correlation matrix 

```python
# Calculating the corr matrix
matrixcorr = Data_Frame2015.corr()
# Creating the heatmap correlation matrix
sns.heatmap(matrixcorr, cmap='PiYG', annot=True)

# Show the plot
plt.show()
```

## 3. Calculating the correlation coefficients

```python
# calculate the correlation coefficients between Happiness Score and each variable

happiness_economy = Data_Frame2015['Happiness Score'].corr(Data_Frame2015['Economy (GDP per Capita)'])
print("Corr. between Happiness Score and Economy :%.3f"% happiness_economy)

happiness_health = Data_Frame2015['Happiness Score'].corr(Data_Frame2015['Health (Life Expectancy)'])
print("Corr. between Happiness Score and Health :%.3f"% happiness_health)

happiness_generosity = Data_Frame2015['Happiness Score'].corr(Data_Frame2015['Generosity'])
print("Corr. between Happiness Score and Generosity:%.3f"% happiness_generosity)
```

```python
## Another way to calculate the correlation coefficients using Pearson

economy_corr, _ = pearsonr(Data_Frame2015['Happiness Score'], Data_Frame2015['Economy (GDP per Capita)'])
print("Pearson corr. for Happiness Score and Economy : %.3f" % economy_corr)

health_corr, _ = pearsonr(Data_Frame2015['Happiness Score'], Data_Frame2015['Health (Life Expectancy)'])
print("Pearson corr. for Happiness Score and Health : %.3f"% health_corr)

generosity_corr, _ = pearsonr(Data_Frame2015['Happiness Score'], Data_Frame2015['Generosity'])
print("Pearson corr. for Happiness Score and Generosity: %.3f " % generosity_corr)
```
